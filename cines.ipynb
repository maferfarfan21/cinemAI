# Análisis bivariable y multivariable:

```python
plt.scatter(df['Nota'], df['Edad'])
plt.xlabel('Nota')
plt.ylabel('Edad')
plt.title('Gráfico de Dispersión')
plt.show()


# Calcular la media
media = df['Edad'].mean()

# Calcular la moda (en caso de que haya múltiplas modas, se devuelve una Serie)
moda = df['Edad'].mode()

# Calcular la varianza
varianza = df['Edad'].var()

# Calcular la desviación estándar
desviacion_estandar = df['Edad'].std()

# Imprimir los resultados
print("Media Edad:", media)
print("Moda Edad:", moda.values)  # Se convierte a una lista de valores si hay múltiples modas
print("Varianza Edad:", varianza)
print("Desviación Estándar Edad:", desviacion_estandar)

```
